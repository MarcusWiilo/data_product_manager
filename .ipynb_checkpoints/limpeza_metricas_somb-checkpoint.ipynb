```python
import pandas as pd 

# você não precisa criar uma função para ler um dataset com pd.read_csv e os parametros certos está ok
def leitura_dados(caminho):
    df = pd.read_csv(caminho)
    return df
```

```python
def creatcolunaNomeUsina(x):
    if x == 1363:
        return "A"
    if x == 875:
        return "B"
    if x == 874:
        return "C"
    if x == 131 :
        return "D"
    if x == 133:
        return "E"
    if x == 64:
        return "A"
```

```python
def CriarRegistroUsina(nome,dia, mes, ano, nomereal):
    return str(dia)+","+str(mes)+","+str(ano)+","+str(nomereal)
          
df_sombra = leitura_dados('data_sombra.csv')

df_sombra['NomeReal'] = df_sombra['nome'].apply(lambda x:creatcolunaNomeUsina(x))
df_sombra['RegistroUsina'] = df_sombra.apply(lambda x: CriarRegistroUsina(x['nome'], x['Dia'], x['Mes'], x['Ano'],  x['NomeReal']), axis=1)

# você está tentanto ddeletar essa coluna de uma forma muito ruim
## com df_sombra.drop(['nome'], axis = 1)
del df_sombra['nome']
del df_sombra['NomeReal']
df_sombra.columns = ['Dia','Mes','Ano','Nome','RegistroUsina']
df_sombra.to_csv("data_somb_final.csv", encoding='utf-8', index=False)
df_sombra
```
